# Read in data

Note that spreadsheet actually goes to column BJ, but the last 11 columns are all empty, which causes the reading-in method to crash. We skip those columns at the point of reading in the data, then add them back in (empty) afterwards as a workaround.

```python
no_grants_df = get_sheet_dfs(
    sheet_id="1kn-ifkB3ZJN8MnsuM2bJmucgQESqmttCjNr4wsiBUK8",
    ranges={"Results": "Sheet1!A2:AY603"} # The
  )['Results']

null_cols = [
    "Has the intervention supported higher skilled, higher paid jobs? If so, why?",
    "Employees at Project Completion (FTE)",
    "At risk jobs still in existence post support - FTE (Jobs Safeguarded)",
    "Tonnes CO2 Reduction",
    "Innovation/R&D Spend (post) for past year (including new product design)",
    "Tonnes Waste Reduction to Landfill",
    "Has resource consumption (electricity, gas, waste, and water) or Greenhouse Gas emissions reduced since you started working with the Programme?",
    "Increased Turnover £",
    "Increased Pre-Tax Profits",
    "Increased TO/Employee",
    "Increased GVA"
]

for col in null_cols:
  no_grants_df[col] = None
```

```python
no_grants_df = wrangle_roadmaps(no_grants_df)
```

# Characterise

```python
roadmaps_characterisation = characterise_df_columnwise(no_grants_df)
output_path = 'outputs/characterisation_no_grants.csv'
if IN_COLAB:
  output_path = dirpath + '/analysis/' + output_path
roadmaps_characterisation.to_csv(output_path)
```
